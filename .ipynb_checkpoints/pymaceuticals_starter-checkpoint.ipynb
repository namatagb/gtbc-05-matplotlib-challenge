## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

merged_results = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="inner")
merged_results
```

```python
# Checking the number of mice.
mice_count = len(merged_results["Mouse ID"].value_counts())
print(f"There are {mice_count} mice in this experiment")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = merged_results.loc[merged_results.duplicated(["Mouse ID", "Timepoint"]),"Mouse ID"]
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID
duplicate_mice_g989 = merged_results.loc[merged_results["Mouse ID"]== 'g989']
duplicate_mice_g989
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merged_results[merged_results["Mouse ID"] != "g989"]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_mice_count = len(clean_df["Mouse ID"].value_counts())
cleaned_mice_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen_df = clean_df.groupby(["Drug Regimen"])

tumor_mean = regimen_df["Tumor Volume (mm3)"].mean().round(2)

tumor_median = regimen_df["Tumor Volume (mm3)"].median().round(2)

tumor_variance = regimen_df["Tumor Volume (mm3)"].var().round(2)

tumor_std = regimen_df["Tumor Volume (mm3)"].std().round(2)

tumor_sem = regimen_df["Tumor Volume (mm3)"].sem().round(2)

table_1_df = pd.DataFrame({
    "Mean": tumor_mean,
    "Median": tumor_median,
    "Variance": tumor_variance,
    "Standard Deviaton": tumor_std,
    "SEM": tumor_sem
})
table_1_df

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

table_summary = clean_df.groupby("Drug Regimen").agg({
    "Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]
}).round(2)
table_summary
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
grouped = clean_df.groupby(["Drug Regimen"])
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
for regimen in regimens:
    s = grouped.get_group(regimen)["Tumor Volume (mm3)"]
    s = s.sort_values()
    vol_data.append(s)
    # add subset 
    
    # Determine outliers using upper and lower bounds
    quartiles = s.quantile([.25,.5,.75])
    lowerq = quartiles.loc[0.25]
    upperq = quartiles.loc[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = s[(s<lower_bound)|(s>upper_bound)]
    print(outliers)
    print(regimen)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.style.use("seaborn")
plt.boxplot(vol_data, labels=regimens);
```
